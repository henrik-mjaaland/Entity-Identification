```python
import boto3

# Create a Comprehend client
comprehend = boto3.client('comprehend')

# Email text
email_text = '''
Subject: Congratulations and Bonus Announcement

Dear [Employee's Name],

I hope this email finds you well. I wanted to take this opportunity to express my heartfelt appreciation for your exceptional contributions and hard work throughout the year. Your dedication and commitment to excellence have significantly impacted our team's success and overall company performance.

I am thrilled to inform you that as a token of our gratitude and recognition of your outstanding performance, you will be receiving a well-deserved performance bonus. I am pleased to announce that your bonus amount is $5,000. This bonus reflects the value we place on your exceptional efforts and the significant impact you have made.

The bonus will be disbursed in your upcoming paycheck, scheduled for [paycheck date]. We hope that this additional financial reward will serve as a testament to our appreciation for your dedication and motivate you to continue achieving great results.

Once again, congratulations on this accomplishment! Your consistent hard work, creativity, and dedication have set a remarkable example for the entire team. We deeply appreciate your contributions and look forward to your continued success within our organization.

If you have any questions or require further information, please feel free to reach out to me or the HR department.

Thank you once again for your outstanding performance and congratulations on this well-deserved bonus!

Best regards,

[Your Name]
[Your Position/Title]
'''

# Detecting entities using Comprehend
response = comprehend.detect_entities(Text=email_text, LanguageCode='en')

# Extracting bonus amount
bonus_amount = None
for entity in response['Entities']:
    if entity['Type'] == 'QUANTITY':
        bonus_amount = entity['Text']
        break

if bonus_amount:
    print(f"Bonus Amount: {bonus_amount}")
else:
    print("No bonus amount found in the email.")
```

```python
import boto3
import time

# Create a Textract client
textract = boto3.client('textract')

# Create an S3 client
s3 = boto3.client('s3')

# Specify the path to the PDF file
pdf_file_path = 'C:/Users/henrik.mjaaland/bonus.pdf'

# Specify the S3 bucket and object key
bucket_name = 'payrolldataextractorbucket'
object_key = 'bonus.pdf'

# Upload the PDF file to S3
s3.upload_file(pdf_file_path, bucket_name, object_key)

# Start the Textract job to extract text from the PDF document
response = textract.start_document_text_detection(
    DocumentLocation={
        'S3Object': {
            'Bucket': bucket_name,
            'Name': object_key
        }
    }
)

# Get the job ID
job_id = response['JobId']

# Retrieve the results of the Textract job
while True:
    response = textract.get_document_text_detection(JobId=job_id)
    status = response['JobStatus']
    
    if status in ['SUCCEEDED', 'FAILED']:
        break
    
    # Wait for the job to complete
    time.sleep(5)

# Process the Textract results
if status == 'SUCCEEDED':
    blocks = response['Blocks']
    extracted_text = ''
    
    for block in blocks:
        if block['BlockType'] == 'LINE':
            # Customize the conditions for extracting specific lines
            if 'bonus' in block['Text'].lower() or 'overtime' in block['Text'].lower():
                extracted_text += block['Text'] + '\n'
    
    print(extracted_text)
else:
    print('Textract job failed.')
```
